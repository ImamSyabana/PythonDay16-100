## Challenge


Now that we've found the major with the highest starting salary, can you write the code to find the following:

* What college major has the highest mid-career salary? How much do graduates with this major earn? (Mid-career is defined as having 10+ years of experience).

* Which college major has the lowest starting salary and how much do graduates earn after university?

* Which college major has the lowest mid-career salary and how much can people expect to earn with this degree? 

```python
# What college major has the highest mid-career salary?
max_mid_carrer_salary = clean_df["Mid-Career Median Salary"].max()
max_mid_carrer_salary_id = clean_df["Mid-Career Median Salary"].idxmax()
clean_df["Undergraduate Major"].loc[max_mid_carrer_salary_id]
```

```python
# How much do graduates with this major earn? (Mid-career is defined as having 10+ years of experience).
max_mid_carrer_salary
```

```python
# Which college major has the lowest starting salary
lowest_Starting_salary =clean_df["Starting Median Salary"].min()
lowest_Starting_salary_idx = clean_df["Starting Median Salary"].idxmin()
clean_df["Undergraduate Major"].loc[lowest_Starting_salary_idx]
```

```python
# how much do graduates earn after university?
lowest_Starting_salary
```

```python
# Which college major has the lowest mid-career salary
lowest_mid_carrer = clean_df["Mid-Career Median Salary"].min()
lowest_mid_carrer_idx = clean_df["Mid-Career Median Salary"].idxmin()

clean_df["Undergraduate Major"].loc[lowest_mid_carrer_idx]
```

```python
# how much can people expect to earn with this degree? 
clean_df.loc[clean_df["Mid-Career Median Salary"].idxmin()]
```

Sorting Values & Adding Columns: Majors with the Most Potential vs Lowest Risk

```python
spread_col = clean_df["Mid-Career 90th Percentile Salary"].subtract(clean_df["Mid-Career 10th Percentile Salary"])
clean_df.insert(1, 'Spread', spread_col)
clean_df.head()
```

Sorting by the Lowest Spread

```python
# To see which degrees have the smallest spread, we can use the .sort_values() method. 
low_risk_major = clean_df.sort_values("Spread")
low_risk_major[["Undergraduate Major", "Spread"]].head()
```

## Challenge
